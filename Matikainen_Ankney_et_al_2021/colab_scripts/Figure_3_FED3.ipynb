```
!pip install seaborn==0.9.0
import pandas as pd
from pandas.plotting import parallel_coordinates
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white")
```

```
# import compiled dataframe
data = pd.DataFrame()
url = "https://raw.githubusercontent.com/KravitzLabDevices/FED3/main/Matikainen_Ankney_et_al_2021/data/Fig3/Meals.csv"
data = pd.read_csv(url, parse_dates = True)
data.head()
```

```
#Fig 3B
colors = ["rebeccapurple", "darkorange"]
variable = "MealsPerDay"
fig, ax = plt.subplots(figsize=(2,4))
sns.barplot(x="Group", y=variable , data=data, capsize=.1, ci=None, palette=colors)
sns.stripplot(x="Group", y=variable , data=data, color="0", alpha=.4, s=12, jitter=False)
sns.despine()

#Draw connected lines
idx0 = 0
idx1 = 1
locs1 = ax.get_children()[idx0].get_offsets()
locs2 = ax.get_children()[idx1].get_offsets()
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2[i, 0]]
    y = [locs1[i, 1], locs2[i, 1]]
    ax.plot(x, y, color='black', alpha=0.3)
```
